```python
r = requests.get(url)
```

```python
soup = bs4.BeautifulSoup(r.text, 'html.parser')
d = soup.find("div", {"class": "pageContent"})
t = d.find("tbody")
```

```python
lines = t.findAll("tr")
```

```python
class message(object):
    def __init__(self):
        self._date = None
        self._serial = None
        self._title = None
        self._url = None
    
    def __repr__(self):
        s = u'{0}: {1}'.format(
            self.serial,
            self.title
        )
        return s.encode('utf-8')
    
    def __unicode__(self):
        return self.__repr__
    
    @property
    def date(self):
        return self._date.strftime('%Y-%m-%d')
    
    @date.setter
    def date(self, d):
        d = d.replace(u'\u200b', '')
        ds = d.split('/')
        self._date = date(int(ds[2]), int(ds[0]), int(ds[1]))
    
    @property
    def serial(self):
        return self._serial
    
    @serial.setter
    def serial(self, s):
        self._serial = s.replace(u'\u200b', '')
    
    @property
    def title(self):
        return self._title
    
    @title.setter
    def title(self, t):
        self._title = t.replace(u'\u200b', '')
    
    @property
    def url(self):
        return self._url
    
    @url.setter
    def url(self, u):
        self._url = u.replace(u'\u200b', '')
        
    def json_feed(self):
        d = {
            "id": self.serial,
            "content_text": self.title,
            "url": self.url,
            "date_published": self.date
        }
        return d
```

```python
message_list = []
for l in lines[1:]:
    cells = l.findAll("td")
    m = message()
    m.serial = cells[0].text
    m.title = cells[1].text
    m.url = 'http://www.public.navy.mil' + cells[1].a['href']
    m.date = cells[2].text
    message_list.append(m)
print(len(message_list))
```
